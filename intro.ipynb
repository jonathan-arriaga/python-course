## Defining functions

```python
def a_function(value1, value2):
    return value1 + value2
```

```python
# Calling the function
some_result = a_function(1,2)
print(some_result)
```

```python
# Calling the function
some_result = a_function('this will definitely ','not fail!')
print(some_result)
```

```python
# Calling the function
some_result = a_function(["okay, ", "so this won't "], ["fail either?"])
print(''.join(some_result))
```

# Classes in Python

```python
class MyClass:    
    def __init__(self, value1, value2):
        self.first_value = value1
        self.second_value = value2
        self.a_constant_value = 3.1416
    
    def some_function(self):
        print(self.a_constant_value)
        
    def print_class_values(self):
        print("First value: {} \n\nSecond value: {} \n\nConstant value: {}".format(self.first_value, self.second_value, self.a_constant_value))
```

```python
a_class_instance = MyClass('Hello', 'World!')
a_class_instance.print_class_values()
```

# Working with some data

```python
# Random value from a normal distribution
x_dist = np.random.randn(10000,1)

print(type(x_dist))
print(x_dist[0:10])

# Histogram
plt.hist(x_dist, bins=100)
plt.title('X Distribution')
plt.xlabel('X Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot
relation_dist = np.random.randn(1000,2)
print(relation_dist[0:10, :])

x = relation_dist[:,0]
y = relation_dist[:,1]
plt.scatter(x, y)
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.show()
```
